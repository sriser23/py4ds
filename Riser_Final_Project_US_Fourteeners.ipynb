There are 96 fourteeners in the United States, all west of the Mississippi River. Colorado has the most (53) of any single state; Alaska is in second place with 29." - Wikipedia

Traffic Low - The low range of estimated visits in the year 2017

### Scraping Data

```python
#Read the Tables in Wikipedia
import pandas as pd
dataframes = pd.read_html('https://en.wikipedia.org/wiki/Fourteener')
```

```python
# Select the second dataframe that is read and assign it to "all14" df
all14 = dataframes[1]
all14.head()
```

### Clean up the dataset

```python
# Drop all units from columns and convert values to numbers
all14['Elevation'] = all14['Elevation'].str[:-2].astype(float)
all14['Prominence'] = all14['Prominence'].str[:-2].astype(float)
all14['Isolation'] = all14['Isolation'].str[:-3].astype(float)
all14.head()
```

```python
# Break out Lat and Long into their own columns
all14['Lat_all'] = all14['Location'].str[27:-15].astype(float)
all14['Long_all'] = all14['Location'].str[-10:-4].astype(float)
all14['Long_all'] = all14['Long_all'] * -1
```

```python
# Remove old Location Column
all14 = all14.drop(['Location'],axis=1)
all14.head()
```

```python
# Clean up "Mountain Peak" column
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('[','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace(']','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('0','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('1','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('2','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('3','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('4','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('5','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('6','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('7','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('8','',-1)
all14['Mountain Peak'] = all14['Mountain Peak'].str.replace('9','',-1)
```

```python
#Clean up "State" column
all14.loc[all14['State'] == 'Alaska British Columbia', 'State'] = 'Alaska'
all14.loc[all14['State'] == 'Alaska Yukon', 'State'] = 'Alaska'
#all14 = all14.replace('Alaska British Columbia','Alaska')
#all14 = all14.replace('Alaska Yukon','Alaska')
```

```python
# Rename Column Titles to make sure reader knows that these figures are using the metric system (vs CO14 in imperial)
# Additionally, add an underscore to Mountain Peak to enable joins later on
all14 = all14.rename(columns={'Elevation': 'Elevation_m', 
                              'Prominence': 'Prominence_m',
                              'Isolation': 'Isolation_km',
                              'Mountain Peak': 'Mountain_Peak',
                              'Mountain Range': 'Mountain_Range'})
```

```python
# Find rows with NaN and correct manually
all14.loc[all14.Isolation_km.isnull()]
```

```python
# Fix Mountain Peaks' NaN in Isolation_km
all14.loc[all14['Mountain_Peak'] == 'Denali', 'Isolation_km'] = 7450
all14.loc[all14['Mountain_Peak'] == 'Mount Whitney', 'Isolation_km'] = 2651
all14.loc[all14['Mountain_Peak'] == 'Mount Elbert', 'Isolation_km'] = 1079
all14.loc[all14['Mountain_Peak'] == 'Mount Rainier', 'Isolation_km'] = 1177
```

```python
# Show metric and imperial
all14['Elevation_ft'] = all14['Elevation_m'] * 3.28084
all14['Isolation_mi'] = all14['Isolation_km'] * 0.621371
```

```python
# Check the dataset and see how it looks
all14.head()
```

## (Task 8) Linear Regression (1 Pt)
### Predicting the State the Mountain Peak is located given the following:
    ## Elevation_ft
    ## Isolation_mi
    
We cannot use Latitude and Longitude Coordinates because that would be considered leakage

```python
#import the model class
from sklearn.naive_bayes import GaussianNB

# Instantiate the model
model2 = GaussianNB()
```

```python
# Separate the input data into features and target variable. Pick features that are not correlated to the target var.

X = all14[['Elevation_ft','Isolation_mi']]
y = all14['State']
```

```python
# Split the data

from sklearn.model_selection import train_test_split
Xtrain2, Xtest2, ytrain2, ytest2 = train_test_split(X,y,random_state=12)
```

```python
# Train the model on the selected features and target

model2.fit(Xtrain2, ytrain2)
model2
```

```python
# Score the model
model2.score(Xtest2,ytest2)

# 79%! not bad!
```

```python
New_Peak = {'Elevation_ft': 20310,
            'Isolation_mi': 4629}
#make a DF with one row: New_Peak
X_new = pd.DataFrame([New_Peak])
model2.predict(X_new)
```

Not bad! The Gaussian model predicted Alaska accurately.
